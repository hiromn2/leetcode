```python
#defaultdict, Counter, most_common, OrderedDict

# Hash map (dict) — your most used tool
from collections import defaultdict
freq = defaultdict(int)       # never raises KeyError, defaults to 0
freq = defaultdict(list)      # defaults to []

# Counter — frequency map in one line
from collections import Counter
c = Counter("aabbbc")         # {'b': 3, 'a': 2, 'c': 1}
c.most_common(2)              # [('b', 3), ('a', 2)]

# Ordered dict — useful for LRU cache problems
from collections import OrderedDict

# Deque — O(1) append and pop from both ends (use instead of list for queues)
from collections import deque
q = deque([1, 2, 3])
q.appendleft(0)               # [0, 1, 2, 3]
q.popleft()                   # O(1), unlike list.pop(0) which is O(n)

# Heap — always a min-heap in Python
import heapq
h = [3, 1, 2]
heapq.heapify(h)              # in-place
heapq.heappush(h, 0)
heapq.heappop(h)              # returns smallest
# Max-heap trick: store negatives
heapq.heappush(h, -val)

# Set — O(1) lookup
seen = set()
seen.add(x)
x in seen                     # True/False
```

```python
# Exercise 2 — Two sum O(n): Given nums = [2, 7, 11, 15] and target = 9, return the indices of two numbers that sum to target. One pass only.

nums = [2, 7, 11, 15]
target = 9

seen = {}
for i, n in enumerate(nums):
    print(f"i is index {i}, n is the value {n}")
def two_sum(nums, target):
    for i, n in enumerate(nums):
        diff = target - n
        if diff in seen:
            return [seen[diff], i]
        seen[n] = i

print(two_sum(nums, target))
```

```python
# Exercise 3 — Group anagrams Given words = ['eat','tea','tan','ate','nat','bat'], group anagrams together.

from collections import defaultdict

words = ['eat','tea','tan','ate','nat','bat']

seen = defaultdict(list)

for i in words:
    t = tuple(sorted(i))
    if t in seen.keys():
        seen[t].append(i)
    else:
        seen[t] = [i]
print(list(seen.values()))



# ALTERNATIVELY
seen = defaultdict(list)

for i in words:
    t = tuple(sorted(i))
    seen[t].append(i)

print(list(seen.values()))

```

```python
2. defaultdict

defaultdict comes from collections.

from collections import defaultdict

Syntax:

d = defaultdict(default_factory)

where default_factory is usually something callable like:

list
int
set
str

Example:

d = defaultdict(list)

This means:

if d[key] already exists, return its value
if d[key] does not exist, automatically create it as list(), which is []

So:

from collections import defaultdict

d = defaultdict(list)

print(d["x"])   # []

Even though "x" was never added before, it does not error. It creates:

{"x": []}

behind the scenes.

3. Why list and not list()?

You write:

defaultdict(list)

not

defaultdict(list())

because defaultdict wants the function itself, not the result of calling it.

list = the constructor/function
list() = an empty list created right now

So defaultdict(list) means:

whenever a missing key appears, call list() for me

Same idea:

defaultdict(int)   # missing keys become 0
defaultdict(set)   # missing keys become set()
```

```python
# Exercise 4 — Max sum subarray of size k Given nums = [2, 1, 5, 1, 3, 2] and k = 3, find the max sum of any contiguous subarray of exactly length k.
#Hint: Fixed window. Each step: add nums[right], subtract nums[right - k].

nums, k = [2, 1, 5, 1, 3, 2], 3
total = sum(nums[:k])
maximum = total 
for i in range(k, len(nums)):
    total = total + nums[i] - nums[i-k]
    if total > maximum:
        maximum = total
print(maximum)
```
