```python
# Exercise 2 — Two sum O(n): Given nums = [2, 7, 11, 15] and target = 9, return the indices of two numbers that sum to target. One pass only.

nums = [2, 7, 11, 15]
target = 9

seen = {}
for i, n in enumerate(nums):
    print(f"i is index {i}, n is the value {n}")
def two_sum(nums, target):
    for i, n in enumerate(nums):
        diff = target - n
        if diff in seen:
            return [seen[diff], i]
        seen[n] = i

print(two_sum(nums, target))
```

```python
# Exercise 3 — Group anagrams Given words = ['eat','tea','tan','ate','nat','bat'], group anagrams together.

from collections import defaultdict

words = ['eat','tea','tan','ate','nat','bat']

seen = defaultdict(list)

for i in words:
    t = tuple(sorted(i))
    if t in seen.keys():
        seen[t].append(i)
    else:
        seen[t] = [i]
print(list(seen.values()))



# ALTERNATIVELY
seen = defaultdict(list)

for i in words:
    t = tuple(sorted(i))
    seen[t].append(i)

print(list(seen.values()))

```

```python
2. defaultdict

defaultdict comes from collections.

from collections import defaultdict

Syntax:

d = defaultdict(default_factory)

where default_factory is usually something callable like:

list
int
set
str

Example:

d = defaultdict(list)

This means:

if d[key] already exists, return its value
if d[key] does not exist, automatically create it as list(), which is []

So:

from collections import defaultdict

d = defaultdict(list)

print(d["x"])   # []

Even though "x" was never added before, it does not error. It creates:

{"x": []}

behind the scenes.

3. Why list and not list()?

You write:

defaultdict(list)

not

defaultdict(list())

because defaultdict wants the function itself, not the result of calling it.

list = the constructor/function
list() = an empty list created right now

So defaultdict(list) means:

whenever a missing key appears, call list() for me

Same idea:

defaultdict(int)   # missing keys become 0
defaultdict(set)   # missing keys become set()
```

```python
# TWO POINTERS
# Exercise 7 — Container with most water
# Given heights = [1,8,6,2,5,4,8,3,7], find two lines forming a container with the most water.
# Hint: area = min(h[l], h[r]) * (r - l). Always move the shorter pointer inward.

def max_water(heights):
    left, right = 0, len(heights) - 1
    best = 0
    while left < right:
        area = min(heights[left], heights[right]) * (right - left)
        best = max(best, area)
        if heights[left] < heights[right]:
            left += 1
        else:
            right -= 1
    return best

print(max_water([1,8,6,2,5,4,8,3,7]))  # 49
```

```python
# Exercise 11 — House robber (space-optimized)
#Given nums = [2, 7, 9, 3, 1], you can't rob adjacent houses. Return the max money.
# Hint: You only need two rolling variables, not the full array.

def rob(nums):
    prev2 = prev1 = 0
    for n in nums:
        prev2, prev1 = prev1, max(prev1, prev2 + n)
    return prev1

print(rob([2, 7, 9, 3, 1]))  # 12  (2+9+1)
```

```python
#BFS / DFS
##Exercise 12 — Number of islands (DFS)
#grid = [['1','1','0'],['0','1','0'],['0','0','1']] — count the islands.
#Hint: For each unvisited '1', DFS to mark all connected land as '0'. Count how many times you trigger DFS.

def num_islands(grid):
    rows, cols = len(grid), len(grid[0])
    count = 0
    def dfs(r, c):
        if not (0<=r<rows and 0<=c<cols): return
        if grid[r][c] != '1': return
        grid[r][c] = '0'
        for dr, dc in [(1,0),(-1,0),(0,1),(0,-1)]:
            dfs(r+dr, c+dc)
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == '1':
                dfs(r, c); count += 1
    return count

print(num_islands([['1','1','0'],['0','1','0'],['0','0','1']]))  # 2
```

```python
#Exercise 13 — Shortest path in a maze (BFS)
#grid = [[0,0,0],[1,1,0],[1,1,0]]. Find the shortest path from top-left to bottom-right. Return -1 if none.
#Hint: BFS guarantees shortest path. Deque of (row, col, distance). Mark visited on enqueue, not dequeue.

from collections import deque

def shortest_path(grid):
    rows, cols = len(grid), len(grid[0])
    if grid[0][0]==1 or grid[-1][-1]==1: return -1
    q = deque([(0, 0, 1)])
    visited = {(0,0)}
    while q:
        r, c, dist = q.popleft()
        if r==rows-1 and c==cols-1: return dist
        for dr, dc in [(1,0),(-1,0),(0,1),(0,-1)]:
            nr, nc = r+dr, c+dc
            if 0<=nr<rows and 0<=nc<cols and grid[nr][nc]==0 and (nr,nc) not in visited:
                visited.add((nr,nc))
                q.append((nr, nc, dist+1))
    return -1

print(shortest_path([[0,0,0],[1,1,0],[1,1,0]]))  # 5
```
