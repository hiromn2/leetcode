```python
#defaultdict, Counter, most_common, OrderedDict

# Hash map (dict) — your most used tool
from collections import defaultdict
freq = defaultdict(int)       # never raises KeyError, defaults to 0
freq = defaultdict(list)      # defaults to []

# Counter — frequency map in one line
from collections import Counter
c = Counter("aabbbc")         # {'b': 3, 'a': 2, 'c': 1}
c.most_common(2)              # [('b', 3), ('a', 2)]

# Ordered dict — useful for LRU cache problems
from collections import OrderedDict

# Deque — O(1) append and pop from both ends (use instead of list for queues)
from collections import deque
q = deque([1, 2, 3])
q.appendleft(0)               # [0, 1, 2, 3]
q.popleft()                   # O(1), unlike list.pop(0) which is O(n)

# Heap — always a min-heap in Python
import heapq
h = [3, 1, 2]
heapq.heapify(h)              # in-place
heapq.heappush(h, 0)
heapq.heappop(h)              # returns smallest
# Max-heap trick: store negatives
heapq.heappush(h, -val)

# Set — O(1) lookup
seen = set()
seen.add(x)
x in seen                     # True/False
```

```python
# Other commands
all(x > 0 for x in nums)   
any(x > 0 for x in nums)  

3 in [1, 2, 3]         # True
3 in {1, 2, 3}         # True, but faster

# Find index of a value
[10, 20, 30].index(20)  # 1  — raises ValueError if not found

# Count occurrences
[1, 2, 2, 3].count(2)   # 2

# Append, remove
arr.append(5)            # add to end
arr.pop()                # remove from end, O(1)
arr.pop(0)               # remove from front, O(n) — use deque instead
arr.insert(i, val)       # insert at index i, O(n)

# Reverse
arr.reverse()            # in-place
arr[::-1]                # returns new reversed list

# Sort
arr.sort()                          # in-place, ascending
arr.sort(reverse=True)              # descending
arr.sort(key=lambda x: abs(x))      # by absolute value
arr.sort(key=lambda x: (x[1], x[0]))# sort by second element, then first

# Sum, min, max over a list
sum([1, 2, 3])          # 6
sum(x**2 for x in arr)  # sum of squares — works with generator
```

```python
# Exercise 2 — Two sum O(n): Given nums = [2, 7, 11, 15] and target = 9, return the indices of two numbers that sum to target. One pass only.

nums = [2, 7, 11, 15]
target = 9

seen = {}
for i, n in enumerate(nums):
    print(f"i is index {i}, n is the value {n}")
def two_sum(nums, target):
    for i, n in enumerate(nums):
        diff = target - n
        if diff in seen:
            return [seen[diff], i]
        seen[n] = i

print(two_sum(nums, target))
```

```python
# Exercise 3 — Group anagrams Given words = ['eat','tea','tan','ate','nat','bat'], group anagrams together.

from collections import defaultdict

words = ['eat','tea','tan','ate','nat','bat']

seen = defaultdict(list)

for i in words:
    t = tuple(sorted(i))
    if t in seen.keys():
        seen[t].append(i)
    else:
        seen[t] = [i]
print(list(seen.values()))



# ALTERNATIVELY
seen = defaultdict(list)

for i in words:
    t = tuple(sorted(i))
    seen[t].append(i)

print(list(seen.values()))

```

```python
2. defaultdict

defaultdict comes from collections.

from collections import defaultdict

Syntax:

d = defaultdict(default_factory)

where default_factory is usually something callable like:

list
int
set
str

Example:

d = defaultdict(list)

This means:

if d[key] already exists, return its value
if d[key] does not exist, automatically create it as list(), which is []

So:

from collections import defaultdict

d = defaultdict(list)

print(d["x"])   # []

Even though "x" was never added before, it does not error. It creates:

{"x": []}

behind the scenes.

3. Why list and not list()?

You write:

defaultdict(list)

not

defaultdict(list())

because defaultdict wants the function itself, not the result of calling it.

list = the constructor/function
list() = an empty list created right now

So defaultdict(list) means:

whenever a missing key appears, call list() for me

Same idea:

defaultdict(int)   # missing keys become 0
defaultdict(set)   # missing keys become set()
```

```python
# Exercise 4 — Max sum subarray of size k Given nums = [2, 1, 5, 1, 3, 2] and k = 3, find the max sum of any contiguous subarray of exactly length k.
#Hint: Fixed window. Each step: add nums[right], subtract nums[right - k].

nums, k = [2, 1, 5, 1, 3, 2], 3
total = sum(nums[:k])
maximum = total 
for i in range(k, len(nums)):
    total = total + nums[i] - nums[i-k]
    if total > maximum:
        maximum = total
print(maximum)
```

```python
# Exercise 5 — Longest substring without repeating chars: Given s = 'abcabcbb', return the length of the longest substring with all unique characters.


s = 'abcabcbb'
biggest = ''
newest = ''

for i in s:
    if i in newest:
        newest = newest[newest.index(i) + 1:] + i
    else:
        newest += i

    if len(newest) > len(biggest):
        biggest = newest

print(biggest)

### Alternatively

for ch in s:
    if ch in newest:
        newest = newest[newest.index(ch) + 1:] + ch
    else:
        newest += ch
    biggest = newest if len(newest) > len(biggest) else biggest

print(biggest)

####

for ch in s:
    if ch in newest:
        newest = newest[newest.index(ch) + 1:] + ch
    else:
        newest += ch
    biggest = max(biggest, newest, key=len)

print(biggest)

#### 

for ch in s:
    if ch in newest:
        newest = newest[newest.index(ch) + 1:]
    newest += ch

    if len(newest) > len(biggest):
        biggest = newest

print(biggest)

#### 
newest = ''
best_len = 0

for ch in s:
    if ch in newest:
        newest = newest[newest.index(ch) + 1:]
    newest += ch
    best_len = max(best_len, len(newest))

print(best_len)


#######
# VERY GOOD

last_seen = {}
left = 0
best_start = 0
best_len = 0

for right, ch in enumerate(s):
    if ch in last_seen and last_seen[ch] >= left:
        left = last_seen[ch] + 1

    last_seen[ch] = right

    if right - left + 1 > best_len:
        best_len = right - left + 1
        best_start = left

print(s[best_start:best_start + best_len])

#####
#Berst 

last_seen = {}
left = 0
best_start = 0
best_len = 0

for right, ch in enumerate(s):
    if ch in last_seen and last_seen[ch] >= left:
        left = last_seen[ch] + 1

    last_seen[ch] = right

    if right - left + 1 > best_len:
        best_len = right - left + 1
        best_start = left

print(s[best_start:best_start + best_len])
```

```python
# TWO POINTERS
# Exercise 7 — Container with most water
# Given heights = [1,8,6,2,5,4,8,3,7], find two lines forming a container with the most water.
# Hint: area = min(h[l], h[r]) * (r - l). Always move the shorter pointer inward.

def max_water(heights):
    left, right = 0, len(heights) - 1
    best = 0
    while left < right:
        area = min(heights[left], heights[right]) * (right - left)
        best = max(best, area)
        if heights[left] < heights[right]:
            left += 1
        else:
            right -= 1
    return best

print(max_water([1,8,6,2,5,4,8,3,7]))  # 49
```
